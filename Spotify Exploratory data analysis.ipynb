```python
# Importing the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# Reading the dataset

tracks = pd.read_csv(r"C:\Users\Kiruthiha\Documents\Data_Analytics_Coursera\Projects_Data_Analytics\Spotify\tracks.csv")
```

```python
# Previewing the dataset

tracks.head()
```

```python
# Finding null values

pd.isnull(tracks).sum()
```

```python
# Information of the dataset

tracks.info()
```

```python
# Viewing top 10 songs with least popularity

least_popular = tracks.sort_values('popularity', ascending = True).head(10)
least_popular
```

```python
# Descriptive statistics of the dataset

tracks.describe().transpose()
```

```python
# Viewing the most popular songs with popularity > 90

most_popular = tracks.query('popularity>90',inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# setting the index to release date

tracks.set_index('release_date', inplace = True)

# Reformatting release date

tracks.index=pd.to_datetime(tracks.index, format = 'mixed')
tracks.head()
```

```python
# Viewing a specific location's data

tracks[['artists']].iloc[20]
```

```python
# Converting the duration from 'milliseconds' to 'seconds'

tracks['duration'] = tracks['duration_ms'].apply(lambda x : round(x/1000))
tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
# Viewing the duration column

tracks.duration.head()
```

```python
# Finding the correlation between the variables 

corr_df = tracks.drop(['key','mode','explicit'], axis = 1).corr(numeric_only = True , method = 'pearson')
plt.figure(figsize = (10,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# sample dataframe to create regression plots

sample_tracks = tracks.sample(int(0.005*len(tracks)))
```

```python
print(len(sample_tracks))
```

```python
# Regression plot for loudness and danceability

plt.figure(figsize=(9,6))
sns.regplot(data = sample_tracks, y= "loudness", x = "danceability", color = "g", line_kws=dict(color="b")).set(title="Loudness vs Danceability")
```

```python
# Regression plot for loudness and energy

plt.figure(figsize=(9,6))
sns.regplot(data = sample_tracks, y= "loudness", x = "energy", color = "g", line_kws=dict(color="b")).set(title="Loudness vs Energy")
```

```python
# Regression plot for popularity and acousticness

plt.figure(figsize=(9,6))
sns.regplot(data = sample_tracks, y= "popularity", x = "acousticness", color = "g", line_kws=dict(color="b")).set(title="popularity vs acousticness")
```

```python
# Creating a new column for release year

tracks['dates']=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
# Visualizing the number of songs per year since 1922

sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year') 
```

```python
# Change in duration of songs over years

total_dr = tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=years, y=total_dr, ax=ax, errwidth = False)
plt.title("Year vs Duration")
plt.xticks(rotation=90)

plt.show()
```

```python
# Analysis of Spotify features

genre = pd.read_csv(r"C:\Users\Kiruthiha\Documents\Data_Analytics_Coursera\Projects_Data_Analytics\Spotify\SpotifyFeatures.csv")
```

```python
genre.head()
```

```python
# Visualizing the track length across various music genres

plt.title('Duration of songs in different Genres')
sns.color_palette('flare', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=genre, errwidth = False)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
# Visualizing top 5 genres

sns.set_style(style = 'darkgrid')
plt.figure(figsize = (9,5))
popular_genres = genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(data = popular_genres, x = 'popularity', y = 'genre', errwidth = False).set(title='Top 5 genres by popularity')
```

1.Dance enthusiasts tend to favor loud songs, contributing to an increased sense of joy.

2.The quantity of songs available on Spotify has grown over time, possibly as a result of technological progress.

4.The duration of songs are shorter in recent years.
